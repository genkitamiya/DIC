```python
# 欠損値の確認
msno.matrix(train, sparkline=False)

plt.show()
```

```python
# 欠損率の表を作成

mis_val = train.isnull().sum()
mis_val_percent = 100 * train.isnull().sum() / len(train)

mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)

mis_val_table = mis_val_table.rename(
    columns = {0 : 'Missing Values', 1 : '% of Total Values'}
)

# 表を欠損率順に並べ替え、欠損値０のサンプルは除外
mis_val_table_sort = mis_val_table[
    mis_val_table.iloc[:,1] != 0].sort_values('% of Total Values', ascending=False).round(1)

print("欠損値を含む列数：{}".format(str(mis_val_table_sort.shape[0])))
print("-----------------")
print(mis_val_table_sort)
```

>欠損値を含む特徴量は120種中67種と、半数以上である。また、そのほとんどはサンプル数の半分以上が欠損している。欠損値を含むデータに有用な特徴量（＝目的変数と相関する）ものも含まれる可能性が高いため、一概に欠損データを除外すべきでは無いと考えられる。

```python
# ラベルごとのサンプル数を表す円グラフ
%matplotlib inline

fig = plt.figure()
ax = fig.add_subplot()

ax.pie(train['TARGET'].value_counts(), autopct='%1.1f%%')
ax.set_title("Distribution of TARGET")

ax.legend(labels=[0, 1])

plt.show()
```

>上記円グラフの通り、本データセットはTARGET == 0に大変偏っており、不均衡データセットだと分かる。

## 【問題4】データ探索
問題3で設定した課題・疑問を解決するためにデータ探索を進めます。考察を伴う表やグラフを 15個以上 作成してください。<br>

<br>

探索過程で新しい疑問点が生まれた場合は、問題3の回答に加えていってください。

1. 欠損値を含む特徴量の有意性を確かめるため、目的変数との相関を調査した。

```python
# TARGETと正の相関が最も強い10種の特徴量を取得
train.corr()['TARGET'].nlargest(10)
```

>最も強い正の相関はDAYS_BIRTHである。ただ、本データセットのDAYS_BIRTHは負数のため、正確には負の相関と思われる。目的変数は１が未返済を示すため、この場合は顧客の年齢が低いほど、未返済も低いことを表す。

```python
# TARGETと負の相関が最も強い10種の特徴量を取得
train.corr()['TARGET'].nsmallest(10)
```

>最も強い負の相関はEXT_SOURCEの1から3が上位に来た。Kaggleのデータ概要には詳細が無く、外部書類による評価データとの記述のみ記載されている。また、欠損率表から確認できた通り、EXT_SOURCEはいずれも欠損値を含む。目的変数との相関を無視できないため、欠損値は除外せず探索を進める。

2. 数値型データに相関の強い特徴量が見つからなかったため、文字型データの相関も調査する。そのためにまず文字データのラベリングを行う。

```python
# LabelEncodingを行う
# ソースコードはKaggleのkernelから取得

from sklearn.preprocessing import LabelEncoder

# Create a label encoder object
le = LabelEncoder()
le_count = 0

# Iterate through the columns
for col in train:
    if train[col].dtype == 'object':
        # If 2 or fewer unique categories
        if len(list(train[col].unique())) <= 2:
            # Train on the training data
            le.fit(train[col])
            # Transform both training and testing data
            train[col] = le.transform(train[col])
            
            # Keep track of how many columns were label encoded
            le_count += 1
            
print('%d columns were label encoded.' % le_count)
```

```python
# One-Hot Encodingを行う
# ソースコードはKaggleのkernelから取得

# one-hot encoding of categorical variables
train = pd.get_dummies(train)

print('Training Features shape: ', train.shape)
```

>One-Hot Encodingによりダミーデータ列が増えたため、合計列は122から243に増加。

```python
#文字データをラベル化した正の相関
train.corr()['TARGET'].nlargest(10)
```

```python
#文字データをラベル化した負の相関
train.corr()['TARGET'].nsmallest(10)
```

>文字型データの中ではCODE_GENDERが目的変数と相関していることが分かる。弱い相関だが男性が正の相関を示し、女性が負の相関を示す　→男性の顧客は未返済の傾向があり、女性は完済の傾向がある。

3. 最も強い相関を示したラベル毎のデータ分布を確認し、データの傾向を調査する<br>

<br>

高い相関を示した`EXT_SOURCE_1`、`EXT_SOURCE_2`、`EXT_SOURCE_3`、`DAYS_BIRTH`、`CODE_GENDER`がどのような分布を示すかを確認するため、確率密度をラベル毎に出力する。

```python
# EXT_SOURCE、DAYS_BIRTH、CODE_GENDERとTARGETを抽出
train_select = train.loc[:, ['TARGET',
                             'DAYS_BIRTH',
                             'CODE_GENDER_M',
                             'CODE_GENDER_F',
                             'EXT_SOURCE_1',
                             'EXT_SOURCE_2',
                             'EXT_SOURCE_3']]

# DAYS_BIRTHを日数から年数（正値）に変換
train_select['DAYS_BIRTH'] = train_select['DAYS_BIRTH'] / -365

train_select.describe()
```

```python
# 特徴量のKDE図を出力
# ソースコードはkaggle kernelより取得
plt.figure(figsize = (21, 12))

for i, source in enumerate(['DAYS_BIRTH', 'CODE_GENDER_M', 'CODE_GENDER_F',
                            'EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']):
    
    plt.subplot(2, 3, i + 1)
    sns.kdeplot(train_select.loc[train_select['TARGET'] == 0, source], label='target == 0')
    sns.kdeplot(train_select.loc[train_select['TARGET'] == 1, source], label='target == 1')
    
    plt.title('Distribution of {} by Target Value'.format(source))
    plt.xlabel('{}'.format(source)); plt.ylabel('Density');
    
plt.tight_layout(h_pad=2.5)
```

>ここで初めてデータに傾向を確認できた。中でも下記は着目すべき点と思われる。<br>
>- 年齢が高い顧客ほど、未返済率が下がる。逆に完済率には年齢が相関しない。
>- 未返済率に性別の差は無いが、女性の方が完済率は高い。
>- EXT_SOURCE_1とEXT_SOURCE_3は共に数値が高いほど完済率が高い傾向にある。特にEXT_SOURCE_3では顕著。

未返済ラベルの年齢分布をヒストグラムで描画する。

```python
# 未返済ラベルの年齢分布
%matplotlib inline

fig = plt.figure()

plt.title("Distribution of default loans by age")
sns.distplot(train_select.loc[train_select['TARGET'] == 1, 'DAYS_BIRTH'], bins=100, kde=False)

plt.show()
```

>KDE図でも確認できた通り未返済顧客は若年層に集中しており、年齢が上がるにつれて未返済が下がっている。

次に性別による返済率を調査する。下に男女別の円グラフを描画した。

```python
# ラベルごとの性別分布を表す円グラフ
%matplotlib inline

fig, axs = plt.subplots(1, 3, figsize=(8,3))

axs[0].pie(train_select['CODE_GENDER_M'].value_counts(),
           autopct='%1.1f%%')
axs[1].pie(train_select.loc[train_select['TARGET'] == 0, 'CODE_GENDER_M'].value_counts(),
           autopct='%1.1f%%')
axs[2].pie(train_select.loc[train_select['TARGET'] == 1, 'CODE_GENDER_M'].value_counts(),
           autopct='%1.1f%%')

axs[0].set_title("Samples by gender")
axs[1].set_title("Completed loans by gender")
axs[2].set_title("Default loans by gender")

fig.legend(labels=['Female', 'Male'], loc='lower center')

plt.tight_layout()
plt.show()
```

>完済ラベルの３分の２は女性であると分かる。しかし、同時にサンプル全数の３分の２もまた女性であるため、このデータだけで女性の方が完済率が高いと結論できない。

4. 特徴量同士がどのような傾向にあるか確認。

```python
# 抽出した特徴量の相関係数を出力
train_select_corr = train_select.corr()

# ヒートマップを出力
sns.heatmap(train_select_corr, annot=True, square='equal')

plt.show()
```

>DAYS_BIRTHとEXT_SOURCE_1の相関係数が0.6と最も高い相関性を示している。（CODE_GENDER_MとCODE_GENDER_Fの相関係数は-1であるが、これはOne-Hot Encodingの影響である）

DAYS_BIRTHとEXT_SOURCE_1との相関を共KDE図で確認。

```python
%matplotlib inline

# 欠損値を除外し、要素数を揃えるため100000行までを指定
plot_data = train_select.copy()
plot_data = plot_data.dropna().loc[:100000, :]

fig, axs = plt.subplots(1, 2, figsize=(15,5))


sns.kdeplot(plot_data.loc[plot_data['TARGET'] == 0, 'DAYS_BIRTH'],
            plot_data.loc[plot_data['TARGET'] == 0, 'EXT_SOURCE_1'],
            ax=axs[0])
sns.kdeplot(plot_data.loc[plot_data['TARGET'] == 1, 'DAYS_BIRTH'],
            plot_data.loc[plot_data['TARGET'] == 1, 'EXT_SOURCE_1'],
            ax=axs[1])

axs[0].set_title("KDE correlation for TARGET == 0")
axs[1].set_title("KDE correlation for TARGET == 1")


plt.tight_layout(w_pad=2.5)
plt.show()
```

>EXT_SOURCE_1とDAYS_BIRTHの相関をラベル別に描画したが、完済および未返済ラベル共に正の相関を示している。

最後に、`sklearn`の`SimpleImputer`クラスを使用してEXT_SOURCEの欠損値の代替を行い、データ全体への影響を調査する。今回はデータの中央値を代替値として扱う。

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy = 'median')

train_impute = train_select.loc[:,['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']].copy()

train_impute = imputer.fit_transform(train_impute)
                                
columns = ['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']
                                
train_impute = pd.DataFrame(train_impute, columns=columns)
train_impute = pd.concat([train_select['TARGET'], train_impute], axis=1)

for i, source in enumerate(columns):
    print("{}の欠損値数：{}".format(source, train_impute[source].isnull().sum()))
```

>欠損値は全て無事代替できた。

```python
# 欠損値を代替したデータの分布を確認
%matplotlib inline

plt.figure(figsize = (15,5))

for i, source in enumerate(['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3']):
    
    plt.subplot(1, 3, i + 1)
    plt.hist(train_impute[source], bins=50)
    
    plt.title('Dist with Imputation of {}'.format(source))
    plt.xlabel('{}'.format(source))
    plt.ylabel('Density')
    
    print("{}に代入した値：{}".format(source, train_select[source].median()))
    
    
plt.tight_layout(h_pad=2.5)
```

>中央値を欠損値に代入した結果、EXT_SOURCE_1およびEXT_SOURCE_3の分布が大きく変わってしまったことが分かる。これらの特徴量は欠損値を多く含んでいたため（それぞれ56.4%と19.8%）、このように分布が崩れてしまったと考えられる。EXT_SOURCE_2の欠損率はわずか0.2%のため、分布が崩れることはなかった。<br>
>
><br>
>
>結論として、欠損値の数値代替はデータ分布の変化を起こす危険性があるため、欠損率の低いデータセットに行うべきと考えられる。
