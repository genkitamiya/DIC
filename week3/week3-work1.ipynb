```python
import numpy as np
from sklearn.datasets import load_iris

iris = load_iris()
print(iris)
print('loading completed')
```

```python
import pandas as pd

X = pd.DataFrame(data=iris.data, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
y = pd.DataFrame(data=iris.target, columns=['Species'])

print('X =\n{}'.format(X))
print('y =\n{}'.format(y))
```

## 【問題6】図の作成
データは可視化して確認することが大切です。次の3つを行ってください。作図のライブラリはMatplotlibの他seabornもよく使われます。<br>

<br>

- ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）
- 特徴量を一つずつ選び、箱ひげ図を使ってラベルごとのデータの分布を可視化する
- 特徴量を一つずつ選び、バイオリン図を使ってラベルごとのデータの分布を可視化する<br>

箱ひげ図とバイオリン図は似た図を表示します。これらはどう違い、それぞれどのような利点があるのかについて説明してください。

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


#ラベルごとのサンプル数を表す円グラフ

fig = plt.figure()
ax = fig.add_subplot()

ax.pie(df['Species'].value_counts(), labels=iris['target_names'], autopct='%1.1f%%')
ax.set_title("Iris samples by species")

plt.show()
```

```python
%matplotlib inline

#matplotlibのboxplot作成

fig = plt.figure(figsize=(15,15))

fig.suptitle("Boxplot grouped by Species", fontsize=20)

ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

#データを種別に分ける
setosa = df[df.Species == 0]
versicolor = df[df.Species == 1]
virginica = df[df.Species == 2]

iris_labels = ["setosa", "versicolor", "virginica"]

ax1.set_title("Sepal Length")
ax2.set_title("Sepal Width")
ax3.set_title("Petal Length")
ax4.set_title("Petal Width")

ax1.boxplot([setosa.sepal_length, versicolor.sepal_length, virginica.sepal_length],
            labels=iris_labels)
ax2.boxplot([setosa.sepal_width, versicolor.sepal_width, virginica.sepal_width],
            labels=iris_labels)
ax3.boxplot([setosa.petal_length, versicolor.petal_length, virginica.petal_length],
            labels=iris_labels)
ax4.boxplot([setosa.petal_width, versicolor.petal_width, virginica.petal_width],
            labels=iris_labels)


plt.show()
```

```python
%matplotlib inline

#seabornのboxplot作成
#matplotlibではデータを各ラベル毎に分ける前処理が必要だったが、seabornでは不要

fig, axs = plt.subplots(2, 2, figsize=(12, 12))

fig.suptitle("Boxplot grouped by Species", fontsize=20)

sns.boxplot(y='sepal_length', x='Species', data=df, ax=axs[0,0])
sns.boxplot(y='sepal_width', x='Species', data=df, ax=axs[0,1])
sns.boxplot(y='petal_length', x='Species', data=df, ax=axs[1,0])
sns.boxplot(y='petal_width', x='Species', data=df, ax=axs[1,1])

#分布も視認するため、stripplotを併せて描画
sns.stripplot(x='Species', y='sepal_length', data=df, jitter=True, color='black', ax=axs[0,0])
sns.stripplot(x='Species', y='sepal_width', data=df, jitter=True, color='black', ax=axs[0,1])
sns.stripplot(x='Species', y='petal_length', data=df, jitter=True, color='black', ax=axs[1,0])
sns.stripplot(x='Species', y='petal_width', data=df, jitter=True, color='black', ax=axs[1,1])

#for文を利用して各グラフのx軸ラベルを種名に変更
#２次元配列をイテレータが読み込めるよう、.flatメソッドを使用
iris_labels = ["setosa", "versicolor", "virginica"]

for ax in axs.flat:
    ax.set_xticklabels(iris_labels)
    ax.set_xlabel(None)

plt.show()
```

>**箱ひげ図の結果**<br>
>全ての特徴量にわたり、versicolorとvirginicaのデータには交差が見られる。特にsepal_widthにおいては有意差が無いと思われる（正確に計算していないので、はっきりと結論できないが…）。他方、setosa他２種と比べても特徴量の違いが顕著である。中でもpetal_lengthとpetal_widthは差が大きく出てる。

```python
%matplotlib inline

#バイオリンプロット作成

fig, axs = plt.subplots(2, 2, figsize=(12, 10))

fig.suptitle("Violinplot grouped by Species", fontsize=16)

setosa = df[df.Species == 0]
versicolor = df[df.Species == 1]
virginica = df[df.Species == 2]

iris_labels = ["setosa", "versicolor", "virginica"]

sns.violinplot(x=df.Species, y=df.sepal_length, ax=axs[0,0])
sns.violinplot(x=df.Species, y=df.sepal_width, ax=axs[0,1])
sns.violinplot(x=df.Species, y=df.petal_length, ax=axs[1,0])
sns.violinplot(x=df.Species, y=df.petal_width, ax=axs[1,1])

for ax in axs.flat:
    ax.set_xticklabels(iris_labels)
    ax.set_xlabel(None)

plt.show()
```

>**箱ひげ図とバイオリン図の違いについて：**<br>
>バイオリン図には箱ひげ図が表す情報（四分位数や最大最小値）に加えて、データの分布も確認できる。分布は箱ひげ図では判らないため、データによってはバイオリン図が適切である。下図は、分布の変化による各プロットの違いが視覚的に分かる。分布が四分位数に偏るにつれ、バイオリン図のピークが変化するが、箱ひげ図に変化は無い。

## 【問題7】特徴量間の関係の確認
特徴量間の関係を見ていきます。次の4つを行ってください。<br>

<br>

- 縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る
- 全ての組み合わせの散布図を一度に表示させる、散布図行列を作る
- ４つの特徴量に関する相関係数行列を作る
- 相関係数行列のヒートマップを作る<br>

<br>

相関係数行列はDataFrameの`corr()`メソッドを使うことで求められます。

```python
%matplotlib inline

#縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る

fig, axs = plt.subplots(2, 3, figsize=(15,10))

sns.scatterplot(x='sepal_length', y='sepal_width', hue='Species', data=df,
                ax=axs[0,0])
sns.scatterplot(x='sepal_length', y='petal_length', hue='Species', data=df,
                ax=axs[0,1])
sns.scatterplot(x='sepal_length', y='petal_width', hue='Species', data=df,
                ax=axs[0,2])
sns.scatterplot(x='sepal_width', y='petal_length', hue='Species', data=df,
                ax=axs[1,0])
sns.scatterplot(x='sepal_width', y='petal_width', hue='Species', data=df,
                ax=axs[1,1])
sns.scatterplot(x='petal_length', y='petal_width', hue='Species', data=df,
                ax=axs[1,2])
  
        
plt.show()

```

```python
#全ての組み合わせの散布図を一度に表示させる、散布図行列を作る
%matplotlib inline

sns.pairplot(df, hue="Species")   

plt.show()
```

>**散布図行列の結果**<br>
>特徴量の組み合わせは6つあるが、いずれの組み合わせでも、setosaのデータが他２種と交差が少ないことが判る。virginicaとversicolorについては、sepal_lengthとsepal_widthの組み合わせを除き、概ね重複が少ない。

```python
#４つの特徴量に関する相関係数行列を作る
res = X.corr()
print(res)
```

```python
%matplotlib inline
#相関係数行列のヒートマップを作る

mask = np.triu(np.ones_like(res, dtype=np.bool))
sns.heatmap(res, mask=mask)

plt.show()
```

>**特徴量相関ヒートマップの結果**<br>
>petal_lengthとpetal_widthの相関が最も高く、次いでpetal_lengthとsepal_length、petal_widthとsepal_lengthの２組み合わせが高い。最も相関が見られない組み合わせ（相関係数が0に近い）はsepal_lengthとsepal_widthである。
