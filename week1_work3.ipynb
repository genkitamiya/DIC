##  【問題】栗まんじゅうが太陽系を覆う日

1個の栗まんじゅうにバイバインを振りかけた場合、太陽系が栗まんじゅうで覆われるのは何分後でしょうか。これを計算するプログラムを作成してください。

栗まんじゅうや太陽系の体積を調査した上で仮定して計算してください。栗まんじゅうの隙間や重力などは考えず、問題を単純化して構いません。

ボールにバイバインを振りかけた時、東京ドームが一杯になる時間など他の状況にも適用できるような汎用的なプログラムを作成してください。また、グラフを表示するなど分かりやすいノートブックを作成してください。

>本問題を解くにあたり、太陽系の体積は太陽から海王星までの距離（＝半径）より計算した。本来太陽系の形は楕円形であるが、本問題では球体として扱い、下記数式を活用した。
> $$ V = 4/3{\pi}r^3 $$

```python
import matplotlib.pyplot as plt
import math

def baibain(vol_item=0, vol_space=0):

    time = 0
    lst_vol_item = [vol_item]
    lst_time = [0]

    while vol_item < vol_space:
        vol_item *= 2
        time += 5
        lst_vol_item.append(vol_item)
        lst_time.append(time)
        
    return time, lst_vol_item, lst_time

#太陽系は栗まんじゅうで覆われるまで
    #すやの栗まんじゅう = 60mm × 40mm × 30mm
    #太陽系の半径（太陽から海王星までの距離） = 4.545e9 km

R_SOLAR = 4.545e12
VOL_KURI = 60/1000 * 40/1000 * 30/1000
VOL_SOLAR = 4/3 * math.pi * R_SOLAR**3

time, lst_vol_item, lst_time = baibain(VOL_KURI, VOL_SOLAR)

%matplotlib inline
plt.title("Kuri-Manju Explosion")
plt.xlabel("Time (min)")
plt.ylabel("Kuri-Manju Volume (m3)")
plt.plot(lst_time, lst_vol_item,'o',ls='--',c='gray',mfc='blue',mec='blue')
plt.grid(ls='--',alpha=0.5)
plt.show()

print("{}分後に太陽系は栗まんじゅうで覆われる。".format(time))


#東京ドームが野球ボールで埋め尽くされるまで
VOL_BASEBALL = 0.08**3 * 4/3 * math.pi
VOL_TOKYODOME = 1240000

time, lst_vol_item, lst_time = baibain(VOL_BASEBALL,VOL_TOKYODOME)

%matplotlib inline
plt.title("Baseball Explosion")
plt.xlabel("Time (min)")
plt.ylabel("Baseball Volume (m3)")
plt.plot(lst_time, lst_vol_item,'o',ls='--',c='gray',mfc='blue',mec='blue')
plt.grid(ls='--',alpha=0.5)
plt.show()

print("{}分後に東京ドームは野球ボールで埋め尽くされる。".format(time))
```
